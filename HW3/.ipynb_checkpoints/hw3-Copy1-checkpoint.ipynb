```python
print(__doc__)


# Code source: Jaques Grobler
# License: BSD 3 clause


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv("mulivariates2.csv") # 這次使用 pandas 的方式讀取 CSV
#print(data)
data.head()

print(data[["X1", "X2"]].values.tolist()) 
inputX = data[["X1", "X2"]].values.tolist() #將 dataframe 變成 list of lists
inputY1 = list(data["Y1"])
inputY2 = list(data["Y2"])
print(inputY1)
#print(data["X1"].tolist())
#print(data["X1"].values)
#for elements in data["X1"]:
#   print(elements)
```

```python
# Create linear regression object
# 選定所要使用的模型
regr = linear_model.LinearRegression()
regr2 = linear_model.LinearRegression()
# Train the model using the training sets
# 執行機器學習
regr.fit(inputX, inputY1)

print('coefficient:' , regr.coef_) #係數 A
print("intercept:", regr.intercept_, end='\n') #截距 C
regr.fit(inputX, inputY2)
print('coefficient:' , regr.coef_) #係數 A
print("intercept:", regr.intercept_, end='\n') #截距 C

```
