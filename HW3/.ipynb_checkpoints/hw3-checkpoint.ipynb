
# Linear Regression Example

This example uses the only the first feature of the `diabetes` dataset, in
order to illustrate a two-dimensional plot of this regression technique. The
straight line can be seen in the plot, showing how linear regression attempts
to draw a straight line that will best minimize the residual sum of squares
between the observed responses in the dataset, and the responses predicted by
the linear approximation.

The coefficients, the residual sum of squares and the variance score are also
calculated.




```python
print(__doc__)


# Code source: Jaques Grobler
# License: BSD 3 clause


import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv("bivariate.csv") # 這次使用 pandas 的方式讀取 CSV
print(type(data))
print(data)
data.head()

print(data[["X1", "X2"]].values.tolist()) 
inputX = data[["X1", "X2"]].values.tolist() #將 dataframe 變成 list of lists
inputY = list(data["Y"])
print(inputY)
#print(data["X1"].tolist())
#print(data["X1"].values)
#for elements in data["X1"]:
#   print(elements)
```

#### rows 裡面有很多欄位，其中 row[0] 儲存可能跟 x 相關的資料

```python
inputX[0:10]
```

```python
# Create linear regression object
# 選定所要使用的模型
regr = linear_model.LinearRegression()

# Train the model using the training sets
# 執行機器學習
regr.fit(inputX, inputY)

```

```python
# 列印所習得的係數和測試誤差
# The coefficients
# y = A*x + b
print('A:' , regr.coef_[0]) #係數 A
print('B:' , regr.coef_[1]) #係數 B
print("C:", regr.intercept_, end='\n') #截距 C
```
