
# Linear Regression Example

This example uses the only the first feature of the `diabetes` dataset, in
order to illustrate a two-dimensional plot of this regression technique. The
straight line can be seen in the plot, showing how linear regression attempts
to draw a straight line that will best minimize the residual sum of squares
between the observed responses in the dataset, and the responses predicted by
the linear approximation.

The coefficients, the residual sum of squares and the variance score are also
calculated.




```python
print(__doc__)


# Code source: Jaques Grobler
# License: BSD 3 clause


import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

import csv
with open("linear.regression.csv", newline='') as csvfile:
    rows = csv.reader(csvfile)
    # dataX = dataY1 = ... = [] 跟 pointer 有 87%像，會出事(變成 alias)
    dataX = [] 
    dataY1 = []
    dataY2 = []
    dataY3 = []
    # myList = list(rows)
    var = 0
    for row in rows:
        var += 1
        print(row)
        if var!= 1:
            dataX.append([float(row[0])])
            dataY1.append(float(row[1]))
            dataY2.append(float(row[2]))
            dataY3.append(float(row[3]))
    print(dataX)
    print(var)
        
```

#### rows 裡面有很多欄位，其中 row[0] 儲存可能跟 x 相關的資料

```python
dataX[0:10]
```

```python
# Split the data into training/testing sets
# 然後我們把 X 的資料切割成訓練資料和測試資料
dataXtrain = dataX[:-20] #一直拿到倒數 20 停止(拋棄後 20 element)
print(len(dataXtrain))
dataXtest = dataX[-20:] #從倒數 20 element 開始拿
print(dataXtest)
print(len(dataXtest))
```

```python
# Split the targets into training/testing sets
# 將 Y 也切割成訓練資料和測試資料
dataY1train = dataY1[:-20]
dataY1test = dataY1[-20:]
dataY2train = dataY2[:-20]
dataY2test = dataY2[-20:]
dataY3train = dataY3[:-20]
dataY3test = dataY3[-20:]
```

```python
# Create linear regression object
# 選定所要使用的模型
regr = linear_model.LinearRegression()
regr2 = linear_model.LinearRegression()
regr3 = linear_model.LinearRegression()

# Train the model using the training sets
# 執行機器學習
regr.fit(dataX, dataY1)
regr2.fit(dataX, dataY2)
regr3.fit(dataX, dataY3)

# Make predictions using the testing set
# 依據所習得的模型，依據測試資料的表徵數值，產生模型所預測的 diabetes 指數
yPred = regr.predict(dataXtest)
```

```python
# 列印所習得的係數和測試誤差
# The coefficients
# y = A*x + b
print('A1: ' , regr.coef_) #係數 A
print("B1: ", regr.intercept_, end='\n') #截距 B

print('\nA2: ' , regr2.coef_)
print("B2: ", regr2.intercept_, end='\n')

print('\nA3: ' , regr2.coef_) #係數 A
print("B3: ", regr2.intercept_, end='\n')
```
